```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.api import add_constant
import pandas as pd
import numpy as np

# Drop the 'place_of_delivery' column as it contains only missing values
if 'place_of_delivery' in df.columns:
    df = df.drop('place_of_delivery', axis=1)

# Mapping for 'Maternal_occupation' column
maternal_occupation_mapping = {'working': 1, 'no work': 0}
df['Maternal_occupation'] = df['Maternal_occupation'].map(maternal_occupation_mapping)

# Fill missing values in 'Maternal_occupation' with the mode, handle empty mode case
mode_maternal_occupation = df['Maternal_occupation'].mode()
if not mode_maternal_occupation.empty:
    df['Maternal_occupation'] = df['Maternal_occupation'].fillna(mode_maternal_occupation[0])
else:
    # Handle the case where mode is empty, perhaps fill with a default value or mean if applicable
    # For now, let's fill with 0 assuming it's a reasonable default if no mode exists.
    df['Maternal_occupation'] = df['Maternal_occupation'].fillna(0)


# Explicitly fill missing values in other relevant columns with their mean
columns_with_missing = ['PrEP', 'total_children_everborn', 'Ever_heard_STIs', 'Ever_testedfor_HIV', 'husband_education']
for col in columns_with_missing:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())

# --- Diagnosis Step: Check for NaNs and Infs before VIF calculation ---
display("Checking for NaNs and Infs before VIF calculation:")
display(df.isnull().sum())
display(np.isinf(df).sum())
# ---------------------------------------------------------------------


# Add a constant column to the DataFrame (required for VIF calculation)
X = add_constant(df)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# Calculate VIF, handling potential errors for constant or linearly dependent features
vif_data["VIF"] = [
    variance_inflation_factor(X.values, i) if X.columns[i] != 'const' else 0
    for i in range(X.shape[1])
]

# Display the VIF results
display("Variance Inflation Factor (VIF) for each feature:")
display(vif_data.sort_values(by='VIF', ascending=False))
```

```python
param_grids = {
    'KNN': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance']
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5, 7]
    },
    'CatBoost': {
        'iterations': [100, 200],
        'learning_rate': [0.01, 0.1],
        'depth': [6, 10]
    },
    'LightGBM': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'num_leaves': [31, 63]
    },
    'Gradient Boosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5, 7]
    }
}

display("Hyperparameter grids for the top 5 models:")
display(param_grids)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
from sklearn.svm import LinearSVC # Import LinearSVC
from catboost import CatBoostClassifier # Import CatBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier



# Initialize the models to be tuned
models_to_tune = {
    'KNN': KNeighborsClassifier(),
    'XGBoost': XGBClassifier(random_state=42, eval_metric="logloss"),
    'CatBoost': CatBoostClassifier(random_state=42, verbose=0),
    'LightGBM': LGBMClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

tuned_models = {}
tuned_results = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1-score": [],
    "ROC AUC": []
}


# Iterate through the models for which parameter grids are defined
for model_name, model in models_to_tune.items():
    if model_name in param_grids:
        param_grid = param_grids[model_name]

        grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='roc_auc', n_jobs=-1)
        grid_search.fit(X_train_resampled_rfe, y_train_resampled)

        tuned_models[model_name] = grid_search

        print(f"Finished tuning {model_name}")
        print(f"Best parameters: {grid_search.best_params_}")
        print(f"Best ROC AUC: {grid_search.best_score_:.4f}")
        print("-" * 30)

        # Evaluate the best model from GridSearchCV on the test data
        best_model = grid_search.best_estimator_
        y_pred = best_model.predict(X_test_rfe)

        tuned_results["Model"].append(model_name)
        tuned_results["Accuracy"].append(accuracy_score(y_test, y_pred))
        tuned_results["Precision"].append(precision_score(y_test, y_pred))
        tuned_results["Recall"].append(recall_score(y_test, y_pred))
        tuned_results["F1-score"].append(f1_score(y_test, y_pred))

        # Handle models without predict_proba or decision_function for ROC AUC
        if hasattr(best_model, "predict_proba"):
            y_proba = best_model.predict_proba(X_test_rfe)[:, 1]
            tuned_results["ROC AUC"].append(roc_auc_score(y_test, y_proba))
        elif hasattr(best_model, "decision_function"):
             y_score = best_model.decision_function(X_test_rfe)
             tuned_results["ROC AUC"].append(roc_auc_score(y_test, y_score))
        else:
            tuned_results["ROC AUC"].append("Not available")
    else:
        print(f"Parameter grid for model '{model_name}' not found in param_grids. Skipping tuning.")


tuned_results_df = pd.DataFrame(tuned_results)
display("Evaluation Metrics Table for Tuned Models:")
display(tuned_results_df)
```
