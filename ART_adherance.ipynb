```python
# Clean and standardize Health_facility names
def standardize_facility(name):
    name = str(name).strip()
    if 'ALERT' in name.upper():
        return 'ALERT'
    elif 'ZEW' in name.upper():
        return 'Zewuditu'
    elif 'YEK' in name.upper():
        return 'Yekatit'
    return name

df_combined['Health_facility'] = df_combined['Health_facility'].apply(standardize_facility)

# Check the unique values to verify the change
print("Standardized Health Facilities:")
print(df_combined['Health_facility'].unique())
print(f"\nValue counts:\n{df_combined['Health_facility'].value_counts()}")
```

```python
import pandas as pd

# Define explicit orders for categorical columns where a natural order exists
ordered_category_maps = {
    'Age': ['18–34', '35–49', '≥50'],
    'BMI': ['Underweight', 'Normal', 'Overweight', 'Obese'],
    'duration_with_hiv_years': ['<5 years', '5–9 years', '≥10 years'],
    'art_duration_years': ['<2 years', '2–5 years', '>5 years'],
    'WHO': ['WHO Stage I', 'WHO Stage II', 'WHO Stage III', 'WHO Stage IV'],
    'NutritionalStatus': ['Undernourished', 'Normal', 'Overweight/Obese'],
    'ARVRegimen': ['First-line', 'Second-line', 'Third-line'],
    'Refill_Category': ['Short refill', 'Multi-month dispensing (MMD)'],
    'AdheranceLevel': ['Good', 'Poor'],
    'viral_load_status': ['Suppressed', 'Unsuppressed'],
    'Opportunistic_Infections': ['No', 'Yes'],
    'TPT_Completed': ['No', 'Yes'],
    'Sex': ['F', 'M'] # Add Sex as an ordered categorical, assuming 'F' then 'M' order
}

# List of all categorical columns for processing
categorical_cols = list(ordered_category_maps.keys()) + [
    # 'Health_facility', # Health_facility will be dropped, no need to process here
    'FunctionalStatus'
]

# Ensure columns are categorical and ordered where defined
for col, order in ordered_category_maps.items():
    if col in df_adult.columns:
        df_adult[col] = pd.Categorical(df_adult[col], categories=order, ordered=True)

# Impute missing values (NaN) in df_adult with the mode for each categorical column
# This loop should run AFTER setting categorical types to ensure mode is calculated correctly
for col in categorical_cols:
    if col in df_adult.columns and df_adult[col].isnull().any():
        # For ordered categories, mode() might return multiple if frequencies are tied.
        # Taking the first one is a common approach.
        col_mode = df_adult[col].mode()[0]
        df_adult[col] = df_adult[col].fillna(col_mode)
        print(f"Imputed missing values in '{col}' with mode: {col_mode}")

descriptive_data = []

for col in categorical_cols:
    if col in df_adult.columns:
        # Use sort=False to preserve the natural order for ordered categorical types
        # For non-ordered categorical types (like FunctionalStatus),
        # value_counts() without sort=False usually sorts by frequency (default), which is fine.
        # If a specific order is needed for these, they should be added to ordered_category_maps.
        counts = df_adult[col].value_counts(dropna=False, sort=False)
        proportions = df_adult[col].value_counts(dropna=False, normalize=True, sort=False) * 100

        first_category_for_col = True
        for category, count in counts.items():
            proportion = proportions[category]
            if first_category_for_col:
                descriptive_data.append([
                    col,
                    str(category),
                    count,
                    f"{proportion:.2f}%"
                ])
                first_category_for_col = False
            else:
                descriptive_data.append([
                    '', # Empty string for subsequent categories within the same variable
                    str(category),
                    count,
                    f"{proportion:.2f}%"
                ])

descriptive_table = pd.DataFrame(descriptive_data, columns=['Variables', 'Categories', 'Frequency', 'Percentage within variable'])

# Set pandas options to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

display(descriptive_table)

# Reset display options to default
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.utils import resample

# Split the BALANCED data for training and validation (from previous cell output)
# X_balanced, y_balanced, X_test, y_test are assumed to be available from previous cells
X_train_bal, X_val_bal, y_train_bal, y_val_bal = train_test_split(
    X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced
)

# Dictionary to store trained models and their predictions/probabilities
models_results = {}

# --- 1. Logistic Regression --- #
print("\n--- Training Logistic Regression ---")
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_bal, y_train_bal)
best_lr_model = lr
y_pred_lr = best_lr_model.predict(X_test)
y_prob_lr = best_lr_model.predict_proba(X_test)[:, 1]
models_results['Logistic Regression'] = {'model': best_lr_model, 'y_pred': y_pred_lr, 'y_prob': y_prob_lr}
print("Logistic Regression trained with default parameters.")

# --- 2. Random Forest --- #
print("\n--- Training Random Forest ---")
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
rf.fit(X_train_bal, y_train_bal)
best_rf_model = rf
y_pred_rf = best_rf_model.predict(X_test)
y_prob_rf = best_rf_model.predict_proba(X_test)[:, 1]
models_results['Random Forest'] = {'model': best_rf_model, 'y_pred': y_pred_rf, 'y_prob': y_prob_rf}
print("Random Forest trained with default parameters.")

# --- 3. LightGBM --- #
print("\n--- Training LightGBM ---")
lgbm = LGBMClassifier(random_state=42, verbose=-1, n_jobs=-1)
lgbm.fit(X_train_bal, y_train_bal)
best_lgbm_model = lgbm
y_pred_lgbm = best_lgbm_model.predict(X_test)
y_prob_lgbm = best_lgbm_model.predict_proba(X_test)[:, 1]
models_results['LightGBM'] = {'model': best_lgbm_model, 'y_pred': y_pred_lgbm, 'y_prob': y_prob_lgbm}
print("LightGBM trained with default parameters.")

# --- 4. XGBoost --- #
print("\n--- Training XGBoost ---")
xgb = XGBClassifier(random_state=42, eval_metric='logloss', use_label_encoder=False, n_jobs=-1)
xgb.fit(X_train_bal, y_train_bal)
best_xgb_model = xgb
y_pred_xgb = best_xgb_model.predict(X_test)
y_prob_xgb = best_xgb_model.predict_proba(X_test)[:, 1]
models_results['XGBoost'] = {'model': best_xgb_model, 'y_pred': y_pred_xgb, 'y_prob': y_prob_xgb}
print("XGBoost trained with default parameters.")

# --- 5. Decision Tree --- #
print("\n--- Training Decision Tree ---")
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_bal, y_train_bal)
best_dt_model = dt
y_pred_dt = best_dt_model.predict(X_test)
y_prob_dt = best_dt_model.predict_proba(X_test)[:, 1]
models_results['Decision Tree'] = {'model': best_dt_model, 'y_pred': y_pred_dt, 'y_prob': y_prob_dt}
print("Decision Tree trained with default parameters.")

# --- 6. MLP (Neural Network) --- #
print("\n--- Training MLP (Neural Network) ---")
mlp = MLPClassifier(max_iter=500, random_state=42)
mlp.fit(X_train_bal, y_train_bal)
best_mlp_model = mlp
y_pred_mlp = best_mlp_model.predict(X_test)
y_prob_mlp = best_mlp_model.predict_proba(X_test)[:, 1]
models_results['MLP (Neural Network)'] = {'model': best_mlp_model, 'y_pred': y_pred_mlp, 'y_prob': y_prob_mlp}
print("MLP trained with default parameters.")


# Function to calculate metrics with 95% Confidence Intervals using bootstrapping
def calculate_metrics_with_ci(y_true, y_pred, y_prob, n_bootstrap=1000):
    metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1 Score': [], 'ROC AUC': []}

    # Ensure y_true, y_pred, y_prob are numpy arrays for consistent indexing
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_prob = np.array(y_prob)

    for i in range(n_bootstrap):
        # Bootstrap sampling (stratified by y_true to maintain class balance)
        indices = resample(np.arange(len(y_true)), stratify=y_true, random_state=i)
        y_true_bs = y_true[indices]
        y_pred_bs = y_pred[indices]
        y_prob_bs = y_prob[indices]

        metrics['Accuracy'].append(accuracy_score(y_true_bs, y_pred_bs))
        metrics['Precision'].append(precision_score(y_true_bs, y_pred_bs, zero_division=0))
        metrics['Recall'].append(recall_score(y_true_bs, y_pred_bs, zero_division=0))
        metrics['F1 Score'].append(f1_score(y_true_bs, y_pred_bs, zero_division=0))

        # ROC AUC requires at least two classes in the sample
        if len(np.unique(y_true_bs)) > 1:
            metrics['ROC AUC'].append(roc_auc_score(y_true_bs, y_prob_bs))
        else:
            metrics['ROC AUC'].append(np.nan) # Append NaN if only one class present

    results = {}
    for metric_name, values in metrics.items():
        clean_values = np.array(values)[~np.isnan(values)] # Remove NaNs from ROC AUC if any
        if len(clean_values) > 0:
            mean_val = np.mean(clean_values)
            lower_ci = np.percentile(clean_values, 2.5)
            upper_ci = np.percentile(clean_values, 97.5)
            results[metric_name] = f"{mean_val:.4f} ({lower_ci:.3f}-{upper_ci:.3f})"
        else:
            results[metric_name] = "N/A"
    return results

# --- Consolidate Metrics in a Table with 95% CI --- #
print("\n--- Model Performance Metrics with 95% Confidence Intervals ---")
final_metrics_data = []
for name, data in models_results.items():
    y_pred = data['y_pred']
    y_prob = data['y_prob']
    ci_metrics = calculate_metrics_with_ci(y_test, y_pred, y_prob)
    ci_metrics['Model'] = name
    final_metrics_data.append(ci_metrics)

metrics_df_ci = pd.DataFrame(final_metrics_data)
metrics_df_ci = metrics_df_ci[['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC']]
print(metrics_df_ci.sort_values(by='ROC AUC', ascending=False).to_string(index=False))

# --- Plot Confusion Matrices (2x3 Grid) --- #
print("\n--- Plotting Confusion Matrices ---")
fig_cm, axes_cm = plt.subplots(2, 3, figsize=(18, 12), dpi=300)
axes_cm = axes_cm.flatten()
class_labels = ['Poor', 'Good']

colors = ['Blues', 'Greens', 'Purples', 'Oranges', 'YlGnBu', 'PuBuGn'] # Color palettes for each model

for i, (name, data) in enumerate(models_results.items()):
    y_pred = data['y_pred']
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap=colors[i % len(colors)], ax=axes_cm[i],
                cbar=False, annot_kws={"size": 14, "weight": "bold"})
    axes_cm[i].set_title(name, fontsize=15, fontweight='bold', pad=12)
    axes_cm[i].set_xticklabels(class_labels, fontsize=12)
    axes_cm[i].set_yticklabels(class_labels, fontsize=12)
    axes_cm[i].set_xlabel('Predicted Label', fontsize=12, fontweight='semibold')
    axes_cm[i].set_ylabel('True Label', fontsize=12, fontweight='semibold')

plt.suptitle('Comparison of Confusion Matrices Across Tuned Models', fontsize=22, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Plot ROC AUC Curves (One Graph) --- #
print("\n--- Plotting ROC AUC Curves ---")
plt.figure(figsize=(12, 8), dpi=300)
sns.set_style("whitegrid")

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier (AUC = 0.50)')

line_colors = ['royalblue', 'forestgreen', 'darkviolet', 'darkorange', 'teal', 'crimson']

for i, (name, data) in enumerate(models_results.items()):
    y_prob = data['y_prob']
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = roc_auc_score(y_test, y_prob)
    plt.plot(fpr, tpr, lw=2.5, color=line_colors[i % len(line_colors)], label=f'{name} (AUC = {roc_auc:.4f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12, fontweight='bold')
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=12, fontweight='bold')
plt.title('Comparison of ROC Curves: All Tuned Models', fontsize=16, pad=20, fontweight='bold')
plt.legend(loc="lower right", fontsize=10, frameon=True, shadow=True)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from scipy.stats import loguniform, randint as sp_randint, uniform as sp_uniform
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.utils import resample

# Split the BALANCED data for training and validation (from previous cell output)
# X_balanced, y_balanced, X_test, y_test are assumed to be available from previous cells
X_train_bal, X_val_bal, y_train_bal, y_val_bal = train_test_split(
    X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced
)

# Dictionary to store best models and their predictions/probabilities
models_results = {}

# --- 1. Logistic Regression --- #
print("\n--- Training and Tuning Logistic Regression ---")
param_dist_lr = {
    'C': loguniform(1e-4, 1e2),
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}
lr = LogisticRegression(max_iter=1000, random_state=42)
random_search_lr = RandomizedSearchCV(
    estimator=lr, param_distributions=param_dist_lr, n_iter=50, cv=5, verbose=0,
    random_state=42, n_jobs=-1, scoring='roc_auc'
)
random_search_lr.fit(X_train_bal, y_train_bal)
best_lr_model = random_search_lr.best_estimator_
y_pred_lr = best_lr_model.predict(X_test)
y_prob_lr = best_lr_model.predict_proba(X_test)[:, 1]
models_results['Logistic Regression'] = {'model': best_lr_model, 'y_pred': y_pred_lr, 'y_prob': y_prob_lr}
print(f"Best LR Params: {random_search_lr.best_params_}")

# --- 2. Random Forest --- #
print("\n--- Training and Tuning Random Forest ---")
param_dist_rf = {
    'n_estimators': sp_randint(100, 500),
    'max_features': ['sqrt', 'log2'],
    'max_depth': sp_randint(10, 100),
    'min_samples_split': sp_randint(2, 20),
    'min_samples_leaf': sp_randint(1, 10),
    'bootstrap': [True, False]
}
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
random_search_rf = RandomizedSearchCV(
    estimator=rf, param_distributions=param_dist_rf, n_iter=50, cv=5, verbose=0,
    random_state=42, n_jobs=-1, scoring='roc_auc'
)
random_search_rf.fit(X_train_bal, y_train_bal)
best_rf_model = random_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
y_prob_rf = best_rf_model.predict_proba(X_test)[:, 1]
models_results['Random Forest'] = {'model': best_rf_model, 'y_pred': y_pred_rf, 'y_prob': y_prob_rf}
print(f"Best RF Params: {random_search_rf.best_params_}")

# --- 3. LightGBM --- #
print("\n--- Training and Tuning LightGBM ---")
param_dist_lgbm = {
    'n_estimators': sp_randint(100, 500),
    'learning_rate': sp_uniform(0.01, 0.3),
    'num_leaves': sp_randint(20, 150),
    'max_depth': sp_randint(5, 50),
    'min_child_samples': sp_randint(10, 100),
    'reg_alpha': sp_uniform(0, 1),
    'reg_lambda': sp_uniform(0, 1)
}
lgbm = LGBMClassifier(random_state=42, verbose=-1, n_jobs=-1)
random_search_lgbm = RandomizedSearchCV(
    estimator=lgbm, param_distributions=param_dist_lgbm, n_iter=50, cv=5, verbose=0,
    random_state=42, n_jobs=-1, scoring='roc_auc'
)
random_search_lgbm.fit(X_train_bal, y_train_bal)
best_lgbm_model = random_search_lgbm.best_estimator_
y_pred_lgbm = best_lgbm_model.predict(X_test)
y_prob_lgbm = best_lgbm_model.predict_proba(X_test)[:, 1]
models_results['LightGBM'] = {'model': best_lgbm_model, 'y_pred': y_pred_lgbm, 'y_prob': y_prob_lgbm}
print(f"Best LGBM Params: {random_search_lgbm.best_params_}")

# --- 4. XGBoost --- #
print("\n--- Training and Tuning XGBoost ---")
param_dist_xgb = {
    'n_estimators': sp_randint(100, 500),
    'learning_rate': sp_uniform(0.01, 0.3),
    'max_depth': sp_randint(3, 15),
    'gamma': sp_uniform(0, 0.5),
    'subsample': sp_uniform(0.6, 0.4),
    'colsample_bytree': sp_uniform(0.6, 0.4),
    'reg_alpha': sp_uniform(0, 1),
    'reg_lambda': sp_uniform(0, 1)
}
xgb = XGBClassifier(random_state=42, eval_metric='logloss', use_label_encoder=False, n_jobs=-1)
random_search_xgb = RandomizedSearchCV(
    estimator=xgb, param_distributions=param_dist_xgb, n_iter=50, cv=5, verbose=0,
    random_state=42, n_jobs=-1, scoring='roc_auc'
)
random_search_xgb.fit(X_train_bal, y_train_bal)
best_xgb_model = random_search_xgb.best_estimator_
y_pred_xgb = best_xgb_model.predict(X_test)
y_prob_xgb = best_xgb_model.predict_proba(X_test)[:, 1]
models_results['XGBoost'] = {'model': best_xgb_model, 'y_pred': y_pred_xgb, 'y_prob': y_prob_xgb}
print(f"Best XGB Params: {random_search_xgb.best_params_}")

# --- 5. Decision Tree --- #
print("\n--- Training and Tuning Decision Tree ---")
param_dist_dt = {
    'max_depth': sp_randint(3, 30),
    'min_samples_split': sp_randint(2, 20),
    'min_samples_leaf': sp_randint(1, 10),
    'criterion': ['gini', 'entropy']
}
dt = DecisionTreeClassifier(random_state=42)
random_search_dt = RandomizedSearchCV(
    estimator=dt, param_distributions=param_dist_dt, n_iter=50, cv=5, verbose=0,
    random_state=42, n_jobs=-1, scoring='roc_auc'
)
random_search_dt.fit(X_train_bal, y_train_bal)
best_dt_model = random_search_dt.best_estimator_
y_pred_dt = best_dt_model.predict(X_test)
y_prob_dt = best_dt_model.predict_proba(X_test)[:, 1]
models_results['Decision Tree'] = {'model': best_dt_model, 'y_pred': y_pred_dt, 'y_prob': y_prob_dt}
print(f"Best DT Params: {random_search_dt.best_params_}")

# --- 6. MLP (Neural Network) --- #
print("\n--- Training and Tuning MLP (Neural Network) ---")
param_dist_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': loguniform(1e-5, 1e-2),
    'learning_rate': ['constant', 'adaptive'],
}
mlp = MLPClassifier(max_iter=500, random_state=42)
random_search_mlp = RandomizedSearchCV(
    estimator=mlp, param_distributions=param_dist_mlp, n_iter=20, cv=3, verbose=0,
    random_state=42, n_jobs=-1, scoring='roc_auc'
)
random_search_mlp.fit(X_train_bal, y_train_bal)
best_mlp_model = random_search_mlp.best_estimator_
y_pred_mlp = best_mlp_model.predict(X_test)
y_prob_mlp = best_mlp_model.predict_proba(X_test)[:, 1]
models_results['MLP (Neural Network)'] = {'model': best_mlp_model, 'y_pred': y_pred_mlp, 'y_prob': y_prob_mlp}
print(f"Best MLP Params: {random_search_mlp.best_params_}")


# Function to calculate metrics with 95% Confidence Intervals using bootstrapping
def calculate_metrics_with_ci(y_true, y_pred, y_prob, n_bootstrap=1000):
    metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1 Score': [], 'ROC AUC': []}

    # Ensure y_true, y_pred, y_prob are numpy arrays for consistent indexing
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_prob = np.array(y_prob)

    for i in range(n_bootstrap):
        # Bootstrap sampling (stratified by y_true to maintain class balance)
        indices = resample(np.arange(len(y_true)), stratify=y_true, random_state=i)
        y_true_bs = y_true[indices]
        y_pred_bs = y_pred[indices]
        y_prob_bs = y_prob[indices]

        metrics['Accuracy'].append(accuracy_score(y_true_bs, y_pred_bs))
        metrics['Precision'].append(precision_score(y_true_bs, y_pred_bs, zero_division=0))
        metrics['Recall'].append(recall_score(y_true_bs, y_pred_bs, zero_division=0))
        metrics['F1 Score'].append(f1_score(y_true_bs, y_pred_bs, zero_division=0))

        # ROC AUC requires at least two classes in the sample
        if len(np.unique(y_true_bs)) > 1:
            metrics['ROC AUC'].append(roc_auc_score(y_true_bs, y_prob_bs))
        else:
            metrics['ROC AUC'].append(np.nan) # Append NaN if only one class present

    results = {}
    for metric_name, values in metrics.items():
        clean_values = np.array(values)[~np.isnan(values)] # Remove NaNs from ROC AUC if any
        if len(clean_values) > 0:
            mean_val = np.mean(clean_values)
            lower_ci = np.percentile(clean_values, 2.5)
            upper_ci = np.percentile(clean_values, 97.5)
            results[metric_name] = f"{mean_val:.4f} ({lower_ci:.3f}-{upper_ci:.3f})"
        else:
            results[metric_name] = "N/A"
    return results

# --- Consolidate Metrics in a Table with 95% CI --- #
print("\n--- Model Performance Metrics with 95% Confidence Intervals ---")
final_metrics_data = []
for name, data in models_results.items():
    y_pred = data['y_pred']
    y_prob = data['y_prob']
    ci_metrics = calculate_metrics_with_ci(y_test, y_pred, y_prob)
    ci_metrics['Model'] = name
    final_metrics_data.append(ci_metrics)

metrics_df_ci = pd.DataFrame(final_metrics_data)
metrics_df_ci = metrics_df_ci[['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC']]
print(metrics_df_ci.sort_values(by='ROC AUC', ascending=False).to_string(index=False))

# --- Plot Confusion Matrices (2x3 Grid) --- #
print("\n--- Plotting Confusion Matrices ---")
fig_cm, axes_cm = plt.subplots(2, 3, figsize=(18, 12), dpi=300)
axes_cm = axes_cm.flatten()
class_labels = ['Poor', 'Good']

colors = ['Blues', 'Greens', 'Purples', 'Oranges', 'YlGnBu', 'PuBuGn'] # Color palettes for each model

for i, (name, data) in enumerate(models_results.items()):
    y_pred = data['y_pred']
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap=colors[i % len(colors)], ax=axes_cm[i],
                cbar=False, annot_kws={"size": 14, "weight": "bold"})
    axes_cm[i].set_title(name, fontsize=15, fontweight='bold', pad=12)
    axes_cm[i].set_xticklabels(class_labels, fontsize=12)
    axes_cm[i].set_yticklabels(class_labels, fontsize=12)
    axes_cm[i].set_xlabel('Predicted Label', fontsize=12, fontweight='semibold')
    axes_cm[i].set_ylabel('True Label', fontsize=12, fontweight='semibold')

plt.suptitle('Comparison of Confusion Matrices Across Tuned Models', fontsize=22, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Plot ROC AUC Curves (One Graph) --- #
print("\n--- Plotting ROC AUC Curves ---")
plt.figure(figsize=(12, 8), dpi=300)
sns.set_style("whitegrid")

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier (AUC = 0.50)')

line_colors = ['royalblue', 'forestgreen', 'darkviolet', 'darkorange', 'teal', 'crimson']

for i, (name, data) in enumerate(models_results.items()):
    y_prob = data['y_prob']
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = roc_auc_score(y_test, y_prob)
    plt.plot(fpr, tpr, lw=2.5, color=line_colors[i % len(line_colors)], label=f'{name} (AUC = {roc_auc:.4f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12, fontweight='bold')
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=12, fontweight='bold')
plt.title('Comparison of ROC Curves: All Tuned Models', fontsize=16, pad=20, fontweight='bold')
plt.legend(loc="lower right", fontsize=10, frameon=True, shadow=True)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
