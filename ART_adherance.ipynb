```python
# Clean and standardize Health_facility names
def standardize_facility(name):
    name = str(name).strip()
    if 'ALERT' in name.upper():
        return 'ALERT'
    elif 'ZEW' in name.upper():
        return 'Zewuditu'
    elif 'YEK' in name.upper():
        return 'Yekatit'
    return name

df_combined['Health_facility'] = df_combined['Health_facility'].apply(standardize_facility)

# Check the unique values to verify the change
print("Standardized Health Facilities:")
print(df_combined['Health_facility'].unique())
print(f"\nValue counts:\n{df_combined['Health_facility'].value_counts()}")
```

```python
# Identify potential date columns
date_columns = [
    'date_hiv_confirmed', 'art_start_date', 'FollowUpDate',
    'CotrimoxazoleStartDate', 'cortimoxazole_stop_date',
    'inhprophylaxis_started_date', 'InhprophylaxisCompletedDate',
    'viral_load_perform_date', 'VL_Sent_Date', 'viral_load_ref_date'
]

# Convert identified columns to datetime
for col in date_columns:
    if col in df_combined.columns:
        df_combined[col] = pd.to_datetime(df_combined[col], dayfirst=True, errors='coerce')

# Verify the conversion
print("Data types for date columns:")
print(df_combined[date_columns].dtypes)
print("\nNumber of successfully parsed dates per column:")
print(df_combined[date_columns].notna().sum())
df_combined.head()
```

```python
# Replace NaN values in AdheranceLevel with 'Poor'
df_combined['AdheranceLevel'] = df_combined['AdheranceLevel'].fillna('Poor')

# Recategorize 'Fair' as 'Poor'
df_combined['AdheranceLevel'] = df_combined['AdheranceLevel'].replace('Fair', 'Poor')

# Calculate missing values (should be 0 now)
missing_count = df_combined['AdheranceLevel'].isna().sum()
total_count = len(df_combined)
proportion_missing = (missing_count / total_count) * 100

print(f"Missing values in 'AdheranceLevel' after replacement: {missing_count}")
print(f"Proportion of missing values: {proportion_missing:.2f}%\n")

# Calculate frequency and proportion of categories within AdheranceLevel
adherence_counts = df_combined['AdheranceLevel'].value_counts(dropna=False)
adherence_proportions = df_combined['AdheranceLevel'].value_counts(normalize=True, dropna=False) * 100

print("Updated AdheranceLevel Categories - Counts and Proportions:")
summary_df = pd.DataFrame({'Count': adherence_counts, 'Proportion (%)': adherence_proportions})
print(summary_df)
```

```python
import pandas as pd

# Define explicit orders for categorical columns where a natural order exists
ordered_category_maps = {
    'Age': ['18–34', '35–49', '≥50'],
    'BMI': ['Underweight', 'Normal', 'Overweight', 'Obese'],
    'duration_with_hiv_years': ['<5 years', '5–9 years', '≥10 years'],
    'art_duration_years': ['<2 years', '2–5 years', '>5 years'],
    'WHO': ['WHO Stage I', 'WHO Stage II', 'WHO Stage III', 'WHO Stage IV'],
    'NutritionalStatus': ['Undernourished', 'Normal', 'Overweight/Obese'],
    'ARVRegimen': ['First-line', 'Second-line', 'Third-line'],
    'Refill_Category': ['Short refill', 'Multi-month dispensing (MMD)'],
    'AdheranceLevel': ['Good', 'Poor'],
    'viral_load_status': ['Suppressed', 'Unsuppressed'],
    'Opportunistic_Infections': ['No', 'Yes'],
    'TPT_Completed': ['No', 'Yes'],
    'Sex': ['F', 'M'] # Add Sex as an ordered categorical, assuming 'F' then 'M' order
}

# List of all categorical columns for processing
categorical_cols = list(ordered_category_maps.keys()) + [
    # 'Health_facility', # Health_facility will be dropped, no need to process here
    'FunctionalStatus'
]

# Ensure columns are categorical and ordered where defined
for col, order in ordered_category_maps.items():
    if col in df_adult.columns:
        df_adult[col] = pd.Categorical(df_adult[col], categories=order, ordered=True)

# Impute missing values (NaN) in df_adult with the mode for each categorical column
# This loop should run AFTER setting categorical types to ensure mode is calculated correctly
for col in categorical_cols:
    if col in df_adult.columns and df_adult[col].isnull().any():
        # For ordered categories, mode() might return multiple if frequencies are tied.
        # Taking the first one is a common approach.
        col_mode = df_adult[col].mode()[0]
        df_adult[col] = df_adult[col].fillna(col_mode)
        print(f"Imputed missing values in '{col}' with mode: {col_mode}")

descriptive_data = []

for col in categorical_cols:
    if col in df_adult.columns:
        # Use sort=False to preserve the natural order for ordered categorical types
        # For non-ordered categorical types (like FunctionalStatus),
        # value_counts() without sort=False usually sorts by frequency (default), which is fine.
        # If a specific order is needed for these, they should be added to ordered_category_maps.
        counts = df_adult[col].value_counts(dropna=False, sort=False)
        proportions = df_adult[col].value_counts(dropna=False, normalize=True, sort=False) * 100

        first_category_for_col = True
        for category, count in counts.items():
            proportion = proportions[category]
            if first_category_for_col:
                descriptive_data.append([
                    col,
                    str(category),
                    count,
                    f"{proportion:.2f}%"
                ])
                first_category_for_col = False
            else:
                descriptive_data.append([
                    '', # Empty string for subsequent categories within the same variable
                    str(category),
                    count,
                    f"{proportion:.2f}%"
                ])

descriptive_table = pd.DataFrame(descriptive_data, columns=['Variables', 'Categories', 'Frequency', 'Percentage within variable'])

# Set pandas options to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

display(descriptive_table)

# Reset display options to default
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.utils import resample

# List of all tuned models
models_tuned = [
    (best_lr_model, "Logistic Regression"),
    (best_rf_model, "Random Forest"),
    (best_lgbm_model, "LightGBM"),
    (best_xgb_model, "XGBoost"),
    (best_dt_model, "Decision Tree"),
    (best_nb_model, "Naive Bayes"),
    (best_knn_model, "KNN"),
    (best_mlp_model, "MLP (Neural Network)")
]

def calculate_metrics_with_ci(model, X, y, n_bootstrap=1000):
    metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1 Score': []}

    # Calculate point estimate for ROC AUC (standard after tuning)
    y_prob = model.predict_proba(X)[:, 1]
    auc_score = roc_auc_score(y, y_prob)

    # Bootstrap for CI of other metrics
    for i in range(n_bootstrap):
        X_bs, y_bs = resample(X, y, stratify=y)
        y_pred_bs = model.predict(X_bs)

        metrics['Accuracy'].append(accuracy_score(y_bs, y_pred_bs))
        metrics['Precision'].append(precision_score(y_bs, y_pred_bs))
        metrics['Recall'].append(recall_score(y_bs, y_pred_bs))
        metrics['F1 Score'].append(f1_score(y_bs, y_pred_bs))

    result = {'ROC AUC': f"{auc_score:.4f}"}
    for m_name, values in metrics.items():
        mean = np.mean(values)
        lower = np.percentile(values, 2.5)
        upper = np.percentile(values, 97.5)
        result[m_name] = f"{mean:.4f} ({lower:.3f}-{upper:.3f})"

    return result

print("Calculating metrics and 95% Confidence Intervals (this may take a moment)...")
final_performance = []
for model, name in models_tuned:
    m_results = calculate_metrics_with_ci(model, X_test, y_test)
    m_results['Model'] = name
    final_performance.append(m_results)

# Create final DataFrame
final_comparison_df = pd.DataFrame(final_performance)
final_comparison_df = final_comparison_df[['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC']]

# Display the table
print("\nFinal Comparison Table - All Models Tuned via RandomizedSearchCV")
print("-----------------------------------------------------------------")
display(final_comparison_df.sort_values(by='ROC AUC', ascending=False).reset_index(drop=True))
```
