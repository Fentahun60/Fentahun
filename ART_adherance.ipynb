```python
# Rename column in df1 to match the other dataframes
df1 = df1.rename(columns={'Health Facility': 'Health_facility'})

# List of dataframes to append
dfs_to_combine = [df1, df2_final, df3_final, df4_final, df5_final]

# Concatenate all dataframes
df_combined = pd.concat(dfs_to_combine, ignore_index=True)

# Display summary info
print(f"Combined DataFrame shape: {df_combined.shape}")
print("Dataframes appended successfully.")
df_combined.head()
```

```python
# Clean and standardize Health_facility names
def standardize_facility(name):
    name = str(name).strip()
    if 'ALERT' in name.upper():
        return 'ALERT'
    elif 'ZEW' in name.upper():
        return 'Zewuditu'
    elif 'YEK' in name.upper():
        return 'Yekatit'
    return name

df_combined['Health_facility'] = df_combined['Health_facility'].apply(standardize_facility)

# Check the unique values to verify the change
print("Standardized Health Facilities:")
print(df_combined['Health_facility'].unique())
print(f"\nValue counts:\n{df_combined['Health_facility'].value_counts()}")
```

```python
df_combined.shape
```

```python
# Calculate proportions of the 'Sex' column
sex_proportions = df_combined['Sex'].value_counts(normalize=True) * 100

print("Proportion of Sex (%):")
print(sex_proportions)

# Also show raw counts for context
print("\nRaw counts:")
print(df_combined['Sex'].value_counts())
```

```python
# Identify potential date columns
date_columns = [
    'date_hiv_confirmed', 'art_start_date', 'FollowUpDate',
    'CotrimoxazoleStartDate', 'cortimoxazole_stop_date',
    'inhprophylaxis_started_date', 'InhprophylaxisCompletedDate',
    'viral_load_perform_date', 'VL_Sent_Date', 'viral_load_ref_date'
]

# Convert identified columns to datetime
for col in date_columns:
    if col in df_combined.columns:
        df_combined[col] = pd.to_datetime(df_combined[col], dayfirst=True, errors='coerce')

# Verify the conversion
print("Data types for date columns:")
print(df_combined[date_columns].dtypes)
print("\nNumber of successfully parsed dates per column:")
print(df_combined[date_columns].notna().sum())
df_combined.head()
```

```python
# Set pandas to display all columns
pd.set_option('display.max_columns', None)

# Display the head of the combined dataframe
df_combined.head()
```

```python
# Replace NaN values in AdheranceLevel with 'Poor'
df_combined['AdheranceLevel'] = df_combined['AdheranceLevel'].fillna('Poor')

# Recategorize 'Fair' as 'Poor'
df_combined['AdheranceLevel'] = df_combined['AdheranceLevel'].replace('Fair', 'Poor')

# Calculate missing values (should be 0 now)
missing_count = df_combined['AdheranceLevel'].isna().sum()
total_count = len(df_combined)
proportion_missing = (missing_count / total_count) * 100

print(f"Missing values in 'AdheranceLevel' after replacement: {missing_count}")
print(f"Proportion of missing values: {proportion_missing:.2f}%\n")

# Calculate frequency and proportion of categories within AdheranceLevel
adherence_counts = df_combined['AdheranceLevel'].value_counts(dropna=False)
adherence_proportions = df_combined['AdheranceLevel'].value_counts(normalize=True, dropna=False) * 100

print("Updated AdheranceLevel Categories - Counts and Proportions:")
summary_df = pd.DataFrame({'Count': adherence_counts, 'Proportion (%)': adherence_proportions})
print(summary_df)
```

```python
from datetime import datetime

# Calculate the duration in years since HIV confirmation and round to 1 decimal place
today = datetime.now()
df_combined['duration_with_hiv_years'] = ((today - df_combined['date_hiv_confirmed']).dt.days / 365.25).round(1)

# Display the first few rows of the relevant columns
print("Summary of the new duration variable (rounded to 1 decimal):")
display(df_combined[['date_hiv_confirmed', 'duration_with_hiv_years']].head())
print(f"\nAverage duration with HIV: {df_combined['duration_with_hiv_years'].mean():.1f} years")
```

```python
# Calculate missing counts and proportions for all variables
missing_counts = df_combined.isna().sum()
missing_proportions = (df_combined.isna().mean() * 100).round(2)

# Create a summary table
missing_table = pd.DataFrame({
    'Missing Count': missing_counts,
    'Proportion (%)': missing_proportions
})

# Display the table
print("Missing Value Analysis for All Variables:")
display(missing_table)
```

```python
# Calculate the mean for Weight and Height
weight_mean = df_combined['Weight'].mean()
height_mean = df_combined['Height'].mean()

# Impute missing values with the calculated means
df_combined['Weight'] = df_combined['Weight'].fillna(weight_mean)
df_combined['Height'] = df_combined['Height'].fillna(height_mean)

# Verify that there are no more missing values in these columns
print(f"Imputed Weight with mean: {weight_mean:.2f}")
print(f"Imputed Height with mean: {height_mean:.2f}")
print(f"Remaining missing values in 'Weight': {df_combined['Weight'].isna().sum()}")
print(f"Remaining missing values in 'Height': {df_combined['Height'].isna().sum()}")
```

```python
# Calculate BMI: Weight(kg) / [Height(m)]^2
# Height is converted from cm to m
df_combined['BMI'] = (df_combined['Weight'] / ((df_combined['Height'] / 100) ** 2)).round(2)

# Display the first few rows of the new variable
print("BMI variable created successfully.")
display(df_combined[['Weight', 'Height', 'BMI']].head())
```

```python
# Define the mapping for standardizing WHO stages
who_mapping = {
    'T1': 'WHO Stage I',
    'Stage One': 'WHO Stage I',
    'T2': 'WHO Stage II',
    'Stage Two': 'WHO Stage II',
    'T3': 'WHO Stage III',
    'Stage Three': 'WHO Stage III',
    'T4': 'WHO Stage IV',
    'Stage Four': 'WHO Stage IV'
}

# Apply the mapping to the 'WHO' column
df_combined['WHO'] = df_combined['WHO'].map(who_mapping).fillna(df_combined['WHO'])

# Recalculate and display the updated distribution
who_counts = df_combined['WHO'].value_counts(dropna=False)
who_proportions = (df_combined['WHO'].value_counts(dropna=False, normalize=True) * 100).round(2)

who_summary = pd.DataFrame({
    'Count': who_counts,
    'Proportion (%)': who_proportions
})

print("Standardized Distribution of WHO Categories:")
display(who_summary)
```

```python
# Define the mapping for recategorizing NutritionalStatus
nutrition_mapping = {
    'Normal': 'Normal',
    'Overweight': 'Overweight/Obese',
    'Obese': 'Overweight/Obese',
    'Mild Malnutrition': 'Undernourished',
    'Moderate Malnutrition': 'Undernourished',
    'Sever Malnutrition': 'Undernourished',
    'SAM': 'Undernourished'
}

# Apply the mapping to the 'NutritionalStatus' column
# fillna(df_combined['NutritionalStatus']) ensures NaNs remain as NaNs
df_combined['NutritionalStatus'] = df_combined['NutritionalStatus'].map(nutrition_mapping).fillna(df_combined['NutritionalStatus'])

# Recalculate and display the updated distribution
nutrition_counts = df_combined['NutritionalStatus'].value_counts(dropna=False)
nutrition_proportions = (df_combined['NutritionalStatus'].value_counts(dropna=False, normalize=True) * 100).round(2)

nutrition_summary = pd.DataFrame({
    'Count': nutrition_counts,
    'Proportion (%)': nutrition_proportions
})

print("Simplified NutritionalStatus Categories:")
display(nutrition_summary)
```

```python
# Define the recategorization function for ARTDoseDays
def categorize_refill(days):
    if pd.isna(days):
        return 'Missing'
    elif days <= 89:
        return 'Short refill'
    elif days >= 90:
        return 'Multi-month dispensing (MMD)'
    return 'Other'

# Apply the recategorization to a new column 'Refill_Category'
df_combined['Refill_Category'] = df_combined['ARTDoseDays'].apply(categorize_refill)

# Calculate frequency and proportion of the new 'Refill_Category'
refill_counts = df_combined['Refill_Category'].value_counts(dropna=False)
refill_proportions = (df_combined['Refill_Category'].value_counts(dropna=False, normalize=True) * 100).round(2)

# Create summary table
refill_summary = pd.DataFrame({
    'Count': refill_counts,
    'Proportion (%)': refill_proportions
})

print("Recategorized Distribution of Refill Category:")
display(refill_summary)
```
