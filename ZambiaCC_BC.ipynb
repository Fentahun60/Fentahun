```python
import pandas as pd
import numpy as np

# Example: your dataframe is df and the column is 'age_at_first_sex'

def recategorize_age(x):
    # Handle non-numeric values first
    if pd.isna(x):
        return np.nan
    if isinstance(x, str) and x.lower() == "not had sex":
        return "not had sex"

    # Convert to numeric if it's a string number
    try:
        x = float(x)
    except:
        return np.nan

    # Categorize
    if x < 15:
        return "<15"
    elif 15 <= x <= 17:
        return "15-17"
    elif 18 <= x <= 19:
        return "18-19"
    elif x >= 20:
        return ">20"
    else:
        return np.nan

# Apply to dataframe
df["age_at_first_sex"] = df["age_at_first_sex"].apply(recategorize_age)

# View result
df["age_at_first_sex"].value_counts(dropna=False)
```

```python
import pandas as pd

# List of variables to tabulate (preserve this order)
variables = [
    'respondent_age', 'residence_type','marital_status', 'highest_edu_level',
    'sex_of_household_head', 'age_of_household_head', 'freq_listen_radio',
    'freq_watch_tv', 'owns_mobile_phone', 'wealth_index',
    'current_contraceptive_method', 'visited_health_facility_12mo',
    'smokes_cigarettes', 'distance_to_health_facility',
    'waiting_time_concern_at_HF', 'covered_by_health_insurance',
    'breast_exam_by_provider', 'age_at_first_sex', 'STI_last_12mo', 'ever_tested_hiv',
    'total_lifetime_sex_partners'
]

# Standardize cervical_cancer_tested to lowercase
df_selected['cervical_cancer_tested'] = df_selected['cervical_cancer_tested'].str.lower()

# Initialize list for results
table_list = []

for var in variables:
    if var == 'cervical_cancer_tested':
        continue  # Skip the outcome variable itself

    # Ensure string/categorical columns are lowercase
    if df_selected[var].dtype == 'object' or str(df_selected[var].dtype) == 'category':
        df_selected[var] = df_selected[var].astype(str).str.lower()

    # Preserve category order if it's categorical
    if hasattr(df_selected[var], 'cat'):
        categories = df_selected[var].cat.categories
    else:
        categories = df_selected[var].unique()

    # Group by variable and cervical_cancer_tested
    counts = df_selected.groupby([var, 'cervical_cancer_tested'], dropna=False).size().unstack(fill_value=0)

    # Append counts in the original category order
    for category in categories:
        if category in counts.index:
            yes_count = counts.loc[category].get('yes', 0)
            no_count = counts.loc[category].get('no', 0)
            table_list.append({
                'Variable': var,
                'Category': category,
                'Yes': yes_count,
                'No': no_count
            })

# Convert to DataFrame
final_table = pd.DataFrame(table_list)

# Display the table
print(final_table)

# Optional: export to CSV
final_table.to_csv("cervical_cancer_table.csv", index=False)
```
