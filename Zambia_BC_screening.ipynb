```python
import pandas as pd
import numpy as np

# Example: your dataframe is df and the column is 'age_at_first_sex'

def recategorize_age(x):
    # Handle non-numeric values first
    if pd.isna(x):
        return np.nan
    if isinstance(x, str) and x.lower() == "not had sex":
        return "not had sex"

    # Convert to numeric if it's a string number
    try:
        x = float(x)
    except:
        return np.nan

    # Categorize
    if x < 15:
        return "<15"
    elif 15 <= x <= 17:
        return "15-17"
    elif 18 <= x <= 19:
        return "18-19"
    elif x >= 20:
        return ">20"
    else:
        return np.nan

# Apply to dataframe
df["age_at_first_sex"] = df["age_at_first_sex"].apply(recategorize_age)

# View result
df["age_at_first_sex"].value_counts(dropna=False)
```

# **Train Test spliting**

```python
from sklearn.model_selection import train_test_split

# Assuming X contains features and y contains the target
# Split data: 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% for testing
    random_state=42,     # for reproducibility
    stratify=y           # keeps class distribution same in train and test
)

# Check the shape
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train distribution:\n", y_train.value_counts())
print("y_test distribution:\n", y_test.value_counts())
```

```python
X_test.columns
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# 1. Initialize Random Forest
# ---------------------------------------------------------
rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight=None
)

# ---------------------------------------------------------
# 2. Apply RFE (Select Top 15 Features)
# ---------------------------------------------------------
n_features_to_select = 15

rfe = RFE(
    estimator=rf,
    n_features_to_select=n_features_to_select,
    step=1
)

rfe.fit(X_train, y_train)

# ---------------------------------------------------------
# 3. Extract ranking results
# ---------------------------------------------------------
ranking_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Rank": rfe.ranking_,
    "Selected": rfe.support_
}).sort_values(by="Rank")

print("\nFeature Ranking (1 = most important):")
print(ranking_df)

# ---------------------------------------------------------
# 4. Create a new dataset with selected features
# ---------------------------------------------------------
selected_features = ranking_df[ranking_df["Selected"]]["Feature"].tolist()

X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

print("\nSelected 15 Features:")
print(selected_features)

# ---------------------------------------------------------
# 5. Publication-Quality Visualization 1: Bar Chart
# ---------------------------------------------------------
plt.style.use("seaborn-v0_8")

plt.figure(figsize=(10, 12), dpi=300)

sorted_df = ranking_df.sort_values("Rank")

bars = plt.barh(
    sorted_df["Feature"], sorted_df["Rank"],
    edgecolor="black", linewidth=1.2,
    color=["#2b83ba" if s else "#c7dcef" for s in sorted_df["Selected"]]
)

plt.gca().invert_yaxis()
plt.title("Feature Ranking using RFE with Random Forest", fontsize=14, fontweight="bold")
plt.xlabel("Rank (1 = Best)", fontsize=14)
plt.ylabel("Features", fontsize=14)
plt.grid(axis='x', linestyle="--", alpha=0.4)
plt.box(False)

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 6. Publication-Quality Visualization 2: Lollipop Plot
# ---------------------------------------------------------
top15 = sorted_df[sorted_df["Selected"]]

plt.figure(figsize=(10, 8), dpi=300)

plt.hlines(
    y=top15["Feature"],
    xmin=0, xmax=top15["Rank"],
    color="#2b83ba", linewidth=2
)

plt.plot(
    top15["Rank"], top15["Feature"],
    "o", markersize=10, color="#2b83ba"
)

plt.gca().invert_yaxis()
plt.xlabel("Rank (1 = Best)", fontsize=14)
plt.title("Top 15 Selected Features (Lollipop Visualization)", fontsize=18, fontweight="bold")
plt.grid(axis='x', linestyle="--", alpha=0.4)
plt.box(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import optuna
from optuna.samplers import TPESampler

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, precision_recall_curve
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# MODELS
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier,
    GradientBoostingClassifier, BaggingClassifier
)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# ------------------------------
# 1. DEFINE MODELS
# ------------------------------
models = {
    "DecisionTree": DecisionTreeClassifier,
    "RandomForest": RandomForestClassifier,
    "ExtraTrees": ExtraTreesClassifier,
    "AdaBoost": AdaBoostClassifier,
    "GradientBoosting": GradientBoostingClassifier,
    "Bagging": BaggingClassifier,
    "LogisticRegression": LogisticRegression,
    "KNN": KNeighborsClassifier,
    "XGBoost": XGBClassifier,
    "LightGBM": LGBMClassifier,
    "CatBoost": CatBoostClassifier
}

# ------------------------------
# 2. TRAIN/VALIDATION SPLIT
# ------------------------------
X_train_opt, X_val, y_train_opt, y_val = train_test_split(
    X_train_res, y_train_res,
    test_size=0.2,
    random_state=42,
    stratify=y_train_res
)

# ------------------------------
# 3. OPTUNA OBJECTIVE FUNCTION
# ------------------------------
def objective(trial, model_name):
    """Objective function optimized by Optuna using F1 score."""

    # Only apply class_weight to models that support it
    weightable_models = ["DecisionTree", "RandomForest", "ExtraTrees", "LogisticRegression", "LightGBM"]

    if model_name == "DecisionTree":
        params = {
            "max_depth": trial.suggest_categorical("max_depth", [3, 5, 10, 15, None]),
            "min_samples_split": trial.suggest_int("min_samples_split", 2, 10),
            "class_weight": "balanced" if model_name in weightable_models else None
        }
        model = DecisionTreeClassifier(**params)

    elif model_name == "RandomForest":
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 200, 500),
            "max_depth": trial.suggest_categorical("max_depth", [5, 10, 20, None]),
            "min_samples_split": trial.suggest_int("min_samples_split", 2, 10),
            "class_weight": "balanced" if model_name in weightable_models else None
        }
        model = RandomForestClassifier(**params)

    elif model_name == "ExtraTrees":
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 200, 500),
            "max_depth": trial.suggest_categorical("max_depth", [5, 10, 20, None]),
            "class_weight": "balanced" if model_name in weightable_models else None
        }
        model = ExtraTreesClassifier(**params)

    elif model_name == "AdaBoost":
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 50, 200),
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 1)
        }
        model = AdaBoostClassifier(**params)

    elif model_name == "GradientBoosting":
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 100, 200),
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
            "max_depth": trial.suggest_int("max_depth", 3, 5)
        }
        model = GradientBoostingClassifier(**params)

    elif model_name == "Bagging":
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 50, 200)
        }
        model = BaggingClassifier(**params)

    elif model_name == "LogisticRegression":
        params = {
            "C": trial.suggest_float("C", 0.1, 10),
            "penalty": "l2",
            "class_weight": "balanced" if model_name in weightable_models else None
        }
        model = LogisticRegression(max_iter=2000, **params)

    elif model_name == "KNN":
        params = {
            "n_neighbors": trial.suggest_int("n_neighbors", 3, 11),
            "weights": trial.suggest_categorical("weights", ["uniform", "distance"])
        }
        model = KNeighborsClassifier(**params)

    elif model_name == "XGBoost":
        pos_weight = (y_train_opt == 0).sum() / (y_train_opt == 1).sum()
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 200, 300),
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
            "max_depth": trial.suggest_int("max_depth", 3, 7),
            "scale_pos_weight": pos_weight
        }
        model = XGBClassifier(eval_metric='logloss', use_label_encoder=False, **params)

    elif model_name == "LightGBM":
        params = {
            "n_estimators": trial.suggest_int("n_estimators", 200, 300),
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
            "num_leaves": trial.suggest_int("num_leaves", 31, 70),
            "class_weight": "balanced"
        }
        model = LGBMClassifier(**params)

    elif model_name == "CatBoost":
        params = {
            "iterations": trial.suggest_int("iterations", 200, 300),
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
            "depth": trial.suggest_int("depth", 4, 8),
            "auto_class_weights": "Balanced"
        }
        model = CatBoostClassifier(verbose=0, **params)

    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", model)
    ])

    pipe.fit(X_train_opt, y_train_opt)
    y_pred = pipe.predict(X_val)
    return f1_score(y_val, y_pred)

# ------------------------------
# 4. TRAIN & TUNE MODELS
# ------------------------------
results = {}
sampler = TPESampler(seed=42)

for name in models.keys():
    print(f"\nðŸ”Ž Tuning {name} with Optuna...")
    study = optuna.create_study(direction="maximize", sampler=sampler)
    study.optimize(lambda trial: objective(trial, name), n_trials=25)

    best_params = study.best_params

    # Handle class weights for applicable models
    if name in ["DecisionTree", "RandomForest", "ExtraTrees", "LogisticRegression", "LightGBM"]:
        best_params["class_weight"] = "balanced"
    if name == "CatBoost":
        best_params["auto_class_weights"] = "Balanced"
    if name == "XGBoost":
        best_params["scale_pos_weight"] = (y_train_res == 0).sum() / (y_train_res == 1).sum()

    best_model = models[name](**best_params)
    pipe = Pipeline([("scaler", StandardScaler()), ("clf", best_model)])
    pipe.fit(X_train_res, y_train_res)

    # Optimize F1 threshold
    y_proba = pipe.predict_proba(X_test_selected)[:, 1]
    precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba)
    f1_scores = 2 * precisions * recalls / (precisions + recalls + 1e-8)
    best_thresh = thresholds[np.argmax(f1_scores)]
    y_pred = (y_proba >= best_thresh).astype(int)

    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC AUC": roc_auc_score(y_test, y_proba),
        "Best Params": best_params,
        "Best Threshold": best_thresh
    }

# ------------------------------
# 5. FINAL RESULTS
# ------------------------------
df_results = pd.DataFrame(results).T
print("\nðŸ“Š FINAL MODEL PERFORMANCE SUMMARY:")
print(df_results)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import shap
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# -----------------------------
# 1. Prepare results for TOPSIS
# -----------------------------
metrics = ["Accuracy", "Precision", "Recall", "F1 Score", "ROC AUC"]
df_metrics = pd.DataFrame({k: {m: v[m] for m in metrics} for k, v in results.items()}).T

# Normalize the metrics (0-1)
scaler = MinMaxScaler()
norm_metrics = pd.DataFrame(scaler.fit_transform(df_metrics), columns=metrics, index=df_metrics.index)

# -----------------------------
# 2. Compute TOPSIS score
# -----------------------------
ideal_best = norm_metrics.max().values
ideal_worst = norm_metrics.min().values

dist_best = np.sqrt(((norm_metrics - ideal_best) ** 2).sum(axis=1))
dist_worst = np.sqrt(((norm_metrics - ideal_worst) ** 2).sum(axis=1))

topsis_score = dist_worst / (dist_best + dist_worst)
df_metrics["TOPSIS"] = topsis_score

# Select best model
best_model_name = df_metrics["TOPSIS"].idxmax()
print(f"ðŸ† Best model by TOPSIS: {best_model_name}")

# -----------------------------
# 3. Visualize TOPSIS scores (with value on each bar)
# -----------------------------
plt.figure(figsize=(10, 6))
bars = plt.bar(df_metrics["TOPSIS"].sort_values(ascending=False).index,
               df_metrics["TOPSIS"].sort_values(ascending=False), color='skyblue')
plt.title("TOPSIS Score for Each Model", fontsize=14)
plt.ylabel("TOPSIS Score")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add score on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.01, f"{height:.3f}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# -----------------------------
# 4. SHAP analysis for the best model
# -----------------------------
best_model_params = results[best_model_name]["Best Params"]
best_model = models[best_model_name](**best_model_params)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", best_model)
])
pipe.fit(X_train_res, y_train_res)

# SHAP explainer
if best_model_name in ["XGBoost", "LightGBM", "CatBoost"]:
    explainer = shap.TreeExplainer(pipe.named_steps["clf"])
else:
    explainer = shap.Explainer(pipe.named_steps["clf"], pipe.named_steps["scaler"].transform(X_train_res))

# Compute SHAP values
shap_values = explainer(pipe.named_steps["scaler"].transform(X_test_selected))

# -----------------------------
# 5. SHAP summary plot (beeswarm)
# -----------------------------
plt.figure(figsize=(12, 8))
shap.summary_plot(shap_values, features=pipe.named_steps["scaler"].transform(X_test_selected),
                  feature_names=X_test_selected.columns, show=True, plot_size=(12, 8))

# -----------------------------
# 6. SHAP bar plot (feature importance)
# -----------------------------
plt.figure(figsize=(10, 6))
shap_values_mean = np.abs(shap_values.values).mean(axis=0)
shap_importance = pd.DataFrame({"Feature": X_test_selected.columns, "SHAP Value": shap_values_mean})
shap_importance = shap_importance.sort_values("SHAP Value", ascending=True)

plt.barh(shap_importance["Feature"], shap_importance["SHAP Value"], color='salmon')
plt.title(f"SHAP Feature Importance - {best_model_name}", fontsize=14)
plt.xlabel("Mean |SHAP Value|")
plt.tight_layout()
plt.show()
```
