```python
variables_to_tabulate = [
    'Region',
    'Maternal_age',
    'highest_educational_level',
    'Residence',
    'sex_of_hh',
    'owns_mobile_telephone',
    'use_of_internet',
    'Wealth_index',
    'currently_pregnant',
    'History_abortion',
    'contraceptive_use_and_intention',
    'visited_HF_last_12months',
    'covered_by_health_insurance',
    'Working_status',
    'Media exposure',
    'gave_birth_in_the_5yrs'
]

# Create an empty list to store the formatted cross-tabulations
formatted_tables = []

for var in variables_to_tabulate:
    # Create the cross-tabulation
    cross_tab = pd.crosstab(df[var], df['Problem_Access'])

    # Calculate row-wise proportions
    cross_tab_prop = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

    # Rename the index to the variable name
    cross_tab.index.name = var
    cross_tab_prop.index.name = var

    # Add a column to identify the variable
    cross_tab['Variable'] = var
    cross_tab_prop['Variable'] = var


    # Reset the index to make the variable categories a column
    cross_tab = cross_tab.reset_index()
    cross_tab_prop = cross_tab_prop.reset_index()


    # Rename the column containing the variable categories
    cross_tab = cross_tab.rename(columns={var: 'Category'})
    cross_tab_prop = cross_tab_prop.rename(columns={var: 'Category'})

    # Format the table to include proportions in brackets
    formatted_cross_tab = cross_tab.copy()
    formatted_cross_tab['big problem'] = formatted_cross_tab['big problem'].astype(str) + ' (' + cross_tab_prop['big problem'].round(2).astype(str) + '%)'
    formatted_cross_tab['not a big problem'] = formatted_cross_tab['not a big problem'].astype(str) + ' (' + cross_tab_prop['not a big problem'].round(2).astype(str) + '%)'


    # Select and reorder columns
    formatted_cross_tab = formatted_cross_tab[['Variable', 'Category', 'big problem', 'not a big problem']]

    # Append to the list
    formatted_tables.append(formatted_cross_tab)

# Concatenate all formatted tables into a single DataFrame
combined_cross_tab = pd.concat(formatted_tables, ignore_index=True)

# Display the combined table
display(combined_cross_tab)

# Save the combined cross-tabulation table to a CSV file
combined_cross_tab.to_csv('combined_cross_tabulation.csv', index=False)

print("\nCombined cross-tabulation table saved as 'combined_cross_tabulation.csv'")
```

```python
# =========================================
# âš™ï¸ Import Libraries
# =========================================
import numpy as np
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import (
    RandomForestClassifier, ExtraTreesClassifier,
    GradientBoostingClassifier, AdaBoostClassifier
)
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier

# =========================================
# ðŸŽ¯ Define Parameter Grids
# =========================================
param_grids = {
    "AdaBoost": {
        "n_estimators": [50, 100, 200, 300],
        "learning_rate": [0.01, 0.05, 0.1, 0.5, 1.0],
    },
    "GradientBoosting": {
        "n_estimators": [100, 200, 300],
        "learning_rate": [0.01, 0.05, 0.1],
        "max_depth": [3, 4, 5],
        "subsample": [0.8, 1.0]
    },
    "XGBoost": {
        "n_estimators": [100, 200, 300],
        "learning_rate": [0.01, 0.05, 0.1],
        "max_depth": [3, 4, 5, 6],
        "subsample": [0.8, 1.0],
        "colsample_bytree": [0.8, 1.0]
    },
    "LightGBM": {
        "n_estimators": [100, 200, 300],
        "learning_rate": [0.01, 0.05, 0.1],
        "num_leaves": [31, 50, 100],
        "subsample": [0.8, 1.0]
    },
    "RandomForest": {
        "n_estimators": [100, 200, 300],
        "max_depth": [None, 10, 20, 30],
        "min_samples_split": [2, 5, 10]
    },
    "ExtraTrees": {
        "n_estimators": [100, 200, 300],
        "max_depth": [None, 10, 20, 30],
        "min_samples_split": [2, 5, 10]
    },
    "DecisionTree": {
        "max_depth": [None, 5, 10, 20, 30],
        "min_samples_split": [2, 5, 10],
        "criterion": ["gini", "entropy"]
    },
    "CatBoost": {
        "iterations": [100, 200, 300],
        "learning_rate": [0.01, 0.05, 0.1],
        "depth": [4, 6, 8]
    }
}

# =========================================
# ðŸš€ Define Models
# =========================================
models = {
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": xgb.XGBClassifier(random_state=42, enable_categorical=True, use_label_encoder=False),
    "LightGBM": lgb.LGBMClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(random_state=42),
    "ExtraTrees": ExtraTreesClassifier(random_state=42),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

# =========================================
# ðŸ§  Feature Selection Handling (RFE)
# =========================================
if 'rfe_selector' not in locals():
    raise ValueError("âŒ RFE selector not found. Please run the RFE feature selection step first.")

# Get selected features
selected_features = X_train.columns[rfe_selector.support_]
print("âœ… Selected Features by RFE:")
display(list(selected_features))

# Subset data
X_train_selected = X_train[selected_features].copy()
X_test_selected = X_test[selected_features].copy()

# Check and convert categorical features if needed
print("\nðŸ“Š Data types before conversion:")
display(X_train_selected.dtypes)

categorical_selected_cols = X_train_selected.select_dtypes(include='category').columns
if len(categorical_selected_cols) > 0:
    print(f"\nðŸ”„ Converting categorical columns to numeric: {list(categorical_selected_cols)}")
    for col in categorical_selected_cols:
        X_train_selected[col] = pd.to_numeric(X_train_selected[col], errors='coerce')
        X_test_selected[col] = pd.to_numeric(X_test_selected[col], errors='coerce')

print("\nâœ… Data types after conversion:")
display(X_train_selected.dtypes)

# =========================================
# ðŸ” Tune Models Using RandomizedSearchCV
# =========================================
best_models = {}

# Identify categorical features for CatBoost
categorical_features_indices = [
    i for i, col in enumerate(X_train_selected.columns)
    if X_train_selected[col].dtype.name == 'category'
]

for name, model in models.items():
    print(f"\nðŸ”¹ Tuning {name}...")
    grid = param_grids[name]

    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=grid,
        n_iter=10,
        scoring="f1",  # you can also use "roc_auc" for balanced datasets
        cv=5,
        verbose=1,
        n_jobs=-1,
        random_state=42
    )

    if name == "CatBoost":
        random_search.fit(X_train_selected, y_train, cat_features=categorical_features_indices)
    else:
        random_search.fit(X_train_selected, y_train)

    best_models[name] = random_search.best_estimator_

    print(f"âœ… Best {name} Parameters: {random_search.best_params_}")
    print(f"ðŸ† Best {name} Score: {random_search.best_score_:.4f}")

print("\nðŸŽ‰ All models tuned and trained successfully!")
```

```python
# =========================================
# ðŸ“Š Model Evaluation
# =========================================
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd

# Dictionary to store evaluation results
evaluation_results = {}

# Evaluate each tuned model on the test set
for model_name, model in best_models.items():
    print(f"\nðŸ” Evaluating {model_name}...")

    # 1ï¸âƒ£ Make predictions
    y_pred = model.predict(X_test_selected)

    # 2ï¸âƒ£ Try to compute ROC-AUC if predict_proba is available
    try:
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_test_selected)[:, 1]  # positive class probability
            roc_auc = roc_auc_score(y_test, y_prob)
        else:
            roc_auc = None
    except Exception:
        roc_auc = None

    # 3ï¸âƒ£ Compute core metrics
    evaluation_results[model_name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, zero_division=0),
        "Recall": recall_score(y_test, y_pred, zero_division=0),
        "F1-Score": f1_score(y_test, y_pred, zero_division=0),
        "ROC-AUC": roc_auc
    }

# 4ï¸âƒ£ Convert to DataFrame for easy comparison
evaluation_df = pd.DataFrame(evaluation_results).T  # models as rows

# 5ï¸âƒ£ Sort models by F1-Score (descending)
evaluation_df = evaluation_df.sort_values(by="F1-Score", ascending=False)

# 6ï¸âƒ£ Display the summary table
print("\nðŸ Model Evaluation Summary:")
display(evaluation_df.style.format({
    "Accuracy": "{:.3f}",
    "Precision": "{:.3f}",
    "Recall": "{:.3f}",
    "F1-Score": "{:.3f}",
    "ROC-AUC": lambda x: "{:.3f}".format(x) if pd.notnull(x) else "N/A"
}).set_table_styles(
    [{"selector": "th", "props": [("text-align", "center")]},
     {"selector": "td", "props": [("text-align", "center")]}]
))

# 7ï¸âƒ£ Identify and display the best model
best_model_name = evaluation_df["F1-Score"].idxmax()
best_model = best_models[best_model_name]
print(f"\nðŸ† Best Model Based on F1-Score: **{best_model_name}**")
```
