```python
variables_to_tabulate = [
    'Region',
    'Maternal_age',
    'highest_educational_level',
    'Residence',
    'sex_of_hh',
    'owns_mobile_telephone',
    'use_of_internet',
    'Wealth_index',
    'currently_pregnant',
    'History_abortion',
    'contraceptive_use_and_intention',
    'visited_HF_last_12months',
    'covered_by_health_insurance',
    'Working_status',
    'Media exposure',
    'gave_birth_in_the_5yrs'
]

# Create an empty list to store the formatted cross-tabulations
formatted_tables = []

for var in variables_to_tabulate:
    # Create the cross-tabulation
    cross_tab = pd.crosstab(df[var], df['Problem_Access'])

    # Calculate row-wise proportions
    cross_tab_prop = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

    # Rename the index to the variable name
    cross_tab.index.name = var
    cross_tab_prop.index.name = var

    # Add a column to identify the variable
    cross_tab['Variable'] = var
    cross_tab_prop['Variable'] = var


    # Reset the index to make the variable categories a column
    cross_tab = cross_tab.reset_index()
    cross_tab_prop = cross_tab_prop.reset_index()


    # Rename the column containing the variable categories
    cross_tab = cross_tab.rename(columns={var: 'Category'})
    cross_tab_prop = cross_tab_prop.rename(columns={var: 'Category'})

    # Format the table to include proportions in brackets
    formatted_cross_tab = cross_tab.copy()
    formatted_cross_tab['big problem'] = formatted_cross_tab['big problem'].astype(str) + ' (' + cross_tab_prop['big problem'].round(2).astype(str) + '%)'
    formatted_cross_tab['not a big problem'] = formatted_cross_tab['not a big problem'].astype(str) + ' (' + cross_tab_prop['not a big problem'].round(2).astype(str) + '%)'


    # Select and reorder columns
    formatted_cross_tab = formatted_cross_tab[['Variable', 'Category', 'big problem', 'not a big problem']]

    # Append to the list
    formatted_tables.append(formatted_cross_tab)

# Concatenate all formatted tables into a single DataFrame
combined_cross_tab = pd.concat(formatted_tables, ignore_index=True)

# Display the combined table
display(combined_cross_tab)

# Save the combined cross-tabulation table to a CSV file
combined_cross_tab.to_csv('combined_cross_tabulation.csv', index=False)

print("\nCombined cross-tabulation table saved as 'combined_cross_tabulation.csv'")
```

```python
# Define ordinal mappings
maternal_age_mapping = {'15-24': 0, '25-34': 1, '35-49': 2}
educational_level_mapping_ordinal = {'No education': 0, 'Primary Education': 1, 'Secondry education and Above': 2, 'Unknown': -1} # Assuming Unknown is treated as missing or a separate category
wealth_index_mapping = {'poorest': 0, 'poorer': 1, 'middle': 2, 'richer': 3, 'richest': 4}

# Apply ordinal encoding in place
df['Maternal_age'] = df['Maternal_age'].map(maternal_age_mapping)
df['highest_educational_level'] = df['highest_educational_level'].map(educational_level_mapping_ordinal)
df['Wealth_index'] = df['Wealth_index'].map(wealth_index_mapping)


# Define binary mappings
problem_access_mapping = {'not a big problem': 0, 'big problem': 1}
residence_mapping = {'rural': 0, 'urban': 1}
sex_of_hh_mapping = {'female': 0, 'male': 1}
owns_mobile_telephone_mapping = {'no': 0, 'yes': 1}
binary_mapping = {'No': 0, 'Yes': 1} # For 'use_of_internet', 'Media exposure', 'gave_birth_in_the_5yrs', 'contraceptive_use_and_intention'
history_abortion_mapping = {'no': 0, 'yes': 1, 9.0: -1} # Assuming 9.0 is unknown/missing
visited_HF_mapping = {'no': 0, 'yes': 1, 9.0: -1} # Assuming 9.0 is unknown/missing
covered_health_insurance_mapping = {'no': 0, 'yes': 1, 9.0: -1} # Assuming 9.0 is unknown/missing
working_status_mapping = {'no': 0, 'yes': 1, 9.0: -1} # Assuming 9.0 is unknown/missing


# Apply binary encoding in place
df['Problem_Access'] = df['Problem_Access'].map(problem_access_mapping)
df['Residence'] = df['Residence'].map(residence_mapping)
df['sex_of_hh'] = df['sex_of_hh'].map(sex_of_hh_mapping)
df['owns_mobile_telephone'] = df['owns_mobile_telephone'].map(owns_mobile_telephone_mapping)
df['use_of_internet'] = df['use_of_internet'].map(binary_mapping)
df['currently_pregnant'] = df['currently_pregnant'].map(binary_mapping) # Assuming currently_pregnant also maps No/Yes
df['History_abortion'] = df['History_abortion'].map(history_abortion_mapping)
df['contraceptive_use_and_intention'] = df['contraceptive_use_and_intention'].map(binary_mapping) # Assuming contraceptive_use_and_intention maps No/Yes
df['visited_HF_last_12months'] = df['visited_HF_last_12months'].map(visited_HF_mapping)
df['covered_by_health_insurance'] = df['covered_by_health_insurance'].map(covered_health_insurance_mapping)
df['Working_status'] = df['Working_status'].map(working_status_mapping)
df['Media exposure'] = df['Media exposure'].map(binary_mapping) # Assuming Media exposure maps No/Yes
df['gave_birth_in_the_5yrs'] = df['gave_birth_in_the_5yrs'].map(binary_mapping) # Assuming gave_birth_in_the_5yrs maps No/Yes


# Columns to keep as they are (including Region and the original identifier columns)
cols_to_keep = ['Country', 'Country_name', 'Region', 'v001', 'v005', 'v007']

# Identify multi-category nominal columns remaining (should be none after binary mapping, but for robustness)
multi_category_nominal_cols = [col for col in df.columns if (df[col].dtype == 'object' or pd.api.types.is_categorical_dtype(df[col])) and col not in cols_to_keep and col not in ordinal_cols and col not in nominal_cols] # Re-evaluate based on current df state

# Perform one-hot encoding for any remaining multi-category nominal variables
df_encoded_multi_nominal = pd.get_dummies(df[multi_category_nominal_cols], columns=multi_category_nominal_cols, drop_first=True)


# Select all columns to include in the final encoded DataFrame
encoded_cols = [col for col in df.columns if col not in multi_category_nominal_cols]


# Combine all encoded and original columns, handling potentially empty df_encoded_multi_nominal
if not df_encoded_multi_nominal.empty:
    df_encoded = pd.concat([df[encoded_cols], df_encoded_multi_nominal], axis=1)
else:
    df_encoded = df[encoded_cols].copy()


# Display the head of the encoded DataFrame
display(df_encoded.head())
```
